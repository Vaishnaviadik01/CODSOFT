### Model Training

```python
from sklearn.model_selection import train_test_split
x=df.drop(columns=['species'])
y=df['species']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
x_test
```

### SVM (Support vector machine algorithm)

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.svm import SVC
model_svc= SVC()
```

```python
model_svc.fit(x_train,y_train)
```

```python
prediction=model_svc.predict(x_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction)*100)
```

```python
model_svc.predict([[6,2,4,1]])
```

### Logistic Regression algorithm

```python
#Instantiate the model
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
```

```python
#model training
model_LR.fit(x_train,y_train)
```

```python
# print metric to get performance.
print('Accuracy:',model_LR.score(x_test,y_test)*100)
```

### KNN Algorithm

```python
#KNN-K-nearest neighbor
from sklearn.neighbors import KNeighborsClassifier
model_knn= KNeighborsClassifier()
```

```python
model_knn.fit(x_train,y_train)
```

```python
# print metric to get performance.
print('Accuracy:',model_knn.score(x_test,y_test)*100)
```

### Dicision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
model_DT= DecisionTreeClassifier()
```

```python
model_DT.fit(x_train,y_train)
```

```python
# print metric to get performance.
print('Accuracy:',model_DT.score(x_test,y_test)*100)
```
